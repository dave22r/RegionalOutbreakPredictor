```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/ca-temp-1525.csv")
df.head()
```

```python
df = df.drop(["STATION", "NAME", "LATITUDE", "LONGITUDE", "ELEVATION"], axis=1)
```

```python
df.head()
```

```python
df["DATE"] = pd.to_datetime(df["DATE"])
df["YEAR"] = df["DATE"].dt.isocalendar().year
df["WEEK"] = df["DATE"].dt.isocalendar().week
df.head()
```

```python
df = df.groupby(["YEAR", "WEEK"]).agg({
    "TAVG": "mean",
    "TMAX": "max",
    "TMIN": "min"
}).reset_index()
df.head()
```

```python
print(df.columns.tolist())
```

```python
df = df[(df['YEAR'] > 2015) | ((df['YEAR'] == 2015) & (df['WEEK'] >= 40))]
df.head()
```

```python
df.to_csv("df-temp-1525.csv", index=False)
```
