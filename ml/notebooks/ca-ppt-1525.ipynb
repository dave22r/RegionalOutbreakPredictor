```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/ca-ppt-1525.csv")
df.head()
```

```python
df = df.drop(["STATION", "NAME", "LATITUDE", "LONGITUDE", "ELEVATION"], axis=1)
```

```python
df["DATE"] = pd.to_datetime(df["DATE"])
df["YEAR"] = df["DATE"].dt.isocalendar().year
df["WEEK"] = df["DATE"].dt.isocalendar().week
df.head()
```

```python
df.tail()
```

```python
df = df.drop(["DATE"], axis=1)
```

```python
column_order = [
    "YEAR",
    "WEEK",
    "PRCP",
]

df = df[column_order]
df.head()
```

```python
df = df.groupby = df.groupby(["YEAR", "WEEK"]).agg({
    "PRCP": "sum"
}).reset_index()
df.head()
```

```python
df.to_csv("df-ppt-1525.csv", index=False)
```
